```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import log_loss,confusion_matrix,classification_report,roc_curve,auc
import pickle
```

```python
#Import Datasets
train=pd.read_csv(r"C:\Users\Vinu\Desktop\Technocolabs\Final Project\train.csv")
test=pd.read_csv(r"C:\Users\Vinu\Desktop\Technocolabs\Final Project\test.csv")
```

```python
target_columns = ['toxic', 'severe_toxic', 'obscene', 'threat','insult', 'identity_hate']
```

```python
s = TfidfVectorizer()
```

```python
X_train = s.fit_transform(train['comment_text'])
X_test = s.transform(test['comment_text'])
```

```python
model=[]
i=0
for label in ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']:
    y = train[label]
    model.append(LogisticRegression(solver='sag'))
    model[i].fit(X_train, y)
    i=i+1
```

```python
y=train[target_columns]
i=0
for col in target_columns:
    print("Column:",col)
    pred_pro = model[i].predict_proba(X_train)[:,1]
    frp,trp,thres = roc_curve(y[col],pred_pro)
    auc_val =auc(frp,trp)
    print(auc_val)
    i=i+1
```

```python
i=0
for col in target_columns:
    test[col]=model[i].predict_proba(X_test)[:, 1]
    i=i+1
test
```

```python
d=open("final_model.pkl", "wb")
pickle.dump(model,d)
pickle.dump(s,d)
```
